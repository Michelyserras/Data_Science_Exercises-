####1 -  Sistema de Categorização numérica.

```python
class Categoria:
  def __init__(self, pares=None, impares=None):
    self.pares = pares if pares is not None else []
    self.impares = impares if impares is not None else []

  def categorizar(self, numeros):
    for num in numeros:
        if num % 2 == 0:
            self.pares.append(num)
        else:
            self.impares.append(num)

  def __repr__(self):
        return f"Categoria(pares={self.pares}, impares={self.impares})"

categoria = Categoria()

numeros = [1,7,10,13,22,4]
categoria.categorizar(numeros)


print(f'Resultados: \nPares: {categoria.pares} Soma: ({sum(categoria.pares)}) \nImpares: {categoria.impares} Soma: ({sum(categoria.impares)})')
print(categoria)
```

####2 -   Sistema de Análise de Números Primos e Compostos.

```python
import math

class Primos:
  def __init__(self, primos=None, compostos=None):
    self.primos = primos if primos is not None else []
    self.compostos = compostos if compostos is not None else []

  def eh_primo(self, numeros):
    for num in numeros:
      primo = True
      if num < 2:
        self.compostos.append(num)
        continue
      for i in range(2, int(math.sqrt(num))+1):
        if num % i == 0:
          primo = False
          break
      if primo:
        self.primos.append(num)
      else:
        self.compostos.append(num)

  def __str__(self):
      return f"Primos: {self.primos} | Quantidade: {len(self.primos)} | Media: {sum(self.primos)/len(self.primos)}\nCompostos: {self.compostos} | Quantidade: {len(self.compostos)} | Media: {sum(self.compostos)/len(self.compostos)}"

cat = Primos()

numeros = [0, 1, 7, 10, 13, 22, 9]

cat.eh_primo(numeros)
print(cat)
```
