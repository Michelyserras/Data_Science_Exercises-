#### Análise de Vendas de Jogos Eletrônicos com Pandas

####Perguntas
1. Qual foi o gênero de jogo com o maior volume de vendas globais (Global_Sales)?
Dica: Use groupby com a coluna Genre e some as vendas globais.

2. Quais são as 5 plataformas com maior média de vendas na América do Norte (NA_Sales)?
Dica: Use groupby na coluna Platform e calcule a média de NA_Sales.

3. Qual publisher lançou o maior número de jogos entre os anos de 2000 e 2010?
Dica: Filtre as linhas com Year entre 2000 e 2010 e conte os jogos por Publisher.

4. Existe alguma relação aparente entre as vendas no Japão (JP_Sales) e o gênero do jogo?
Dica: Compare as médias ou medianas de JP_Sales por Genre e descreva o que observou.

```python
import pandas as pd
import numpy as np
import math

from google.colab import drive
drive.mount('/content/drive')

path = "/content/drive/MyDrive/DataScience"
df = pd.read_csv(path + '/vgsales.csv', sep=',')
df.head(10)

df

```

```python
#ex 1

vendasPorGenero= df.groupby("Genre")["Global_Sales"].sum()

genero_com_maior_venda = vendasPorGenero.idxmax()
maiorVenda = vendasPorGenero.max()

print(f"O genero com maior volume de vendas globais é {genero_com_maior_venda} com um total de {maiorVenda} milhões de cópias vendidas")
```

```python
#ex 2
mediaPorPlataforma = df.groupby("Platform")["NA_Sales"].mean()
maiorPlataformas = mediaPorPlataforma.nlargest(5)

print(f"As plataformas com as maiores medias são: \n {maiorPlataformas} ")
```

```python
#ex 3

jogosEntre2000_2010 = df[(df['Year'] >= 2000) & (df['Year'] <= 2010)]

jogosPorPublisher = jogosEntre2000_2010["Publisher"].value_counts()

pubComMaisJogos = jogosPorPublisher.idxmax()
maiorQtdJogos = jogosPorPublisher.max()

print(f"O publisher que lançou mais jogos entre 2000 e 2010 foi {pubComMaisJogos}, com {maiorQtdJogos} jogos.")
```

```python
#ex 4
mediaJpnGenero = df.groupby("Genre")["JP_Sales"].mean()

medianaJpnGenero = df.groupby("Genre")["JP_Sales"].median()

print(f"Media de JP_SALES por genêro: {mediaJpnGenero} \n\n\nMediana de JP_SALES por genêro {medianaJpnGenero}")
```
